```python
import util
import model

import tensorflow as tf
from tensorflow.keras.callbacks import Callback
import numpy as np
import imgaug as ia

import os
from datetime import datetime
```

```python
# Parameters
params = {'dim': (224,224),
          'batch_size': 4,
          'n_classes': 1048,
          'n_channels': 3,
          'shuffle': True}

# Datasets
partition, labels_tr, classes = util.load_data()

# Generators
image_dir_tr = os.path.join(os.getcwd(), os.pardir, 'data/local_subset/')
gen_tr = model.DataGenerator(partition['train'], labels_tr, image_dir_tr, **params)
gen_val = model.DataGenerator(partition['validation'], labels_tr, image_dir_tr, **params)
```

```python
# Check generator
X, y = gen_tr.__getitem__(0)
for i, img in enumerate(X):
    ia.imshow(img)
#     print(img)
    print(y[i])
    break
```

```python
# Training parameters
lr = .01
batch_size = 4
epochs = 10
logs_dir = os.path.join(os.getcwd(), 'logs/')
first_train = False

# Model
nn = model.create_keras_model(input_shape=(*params['dim'], params['n_channels']),
                              learning_rate=lr)

# Callbacks
lr_decay_cb = tf.keras.callbacks.LearningRateScheduler(
    lambda epoch: lr + 0.02 * (0.5 ** (1 + epoch)),
    verbose=True)

if first_train:
    tensorboard_cb = tf.keras.callbacks.TensorBoard(logs_dir, histogram_freq=1)

class MetricsHistory(Callback):
    def __init__(self):
        self.best_score = -1
    
    def on_batch_end(self, batch, logs={}):
        print()
        print("......Average biggest logit = {:2f}".format(batch, logs["biggest_logit"]))
    
    def on_epoch_end(self, epoch, logs={}):
        curr_score = logs['accuracy_ml']
        print()
        print('Epoch accuracy: {:2f}'.format(curr_score))
        print('Biggest logit: {:2f}'.format(logs["biggest_logit"]))
        
        if curr_score > self.best_score:
            print("Better score, saving model ...")
            self.model.save(logs_dir+'best_model-'+ datetime.now().strftime("%m-%d-%H") +'.h5')
            self.best_score = curr_score

metrics_cb = MetricsHistory()
# TODO: CSVLogger(logs_dir+'train_log.csv')
```

```python
if not first_train:
    !rm -rf logs
%tensorboard --logdir logs/
```

```python
# Train
history = nn.fit_generator(gen_tr, epochs=epochs, validation_data=gen_val,
                 callbacks=[lr_decay_cb, tensorboard_cb, metrics_cb])
# TODO: use_multiprocessing=True, workers=6, validation_steps=val_steps, 
```

```python
# Export
export_path = os.path.join(logs_dir, 'keras_export')
tf.keras.models.save_model(nn, export_path)
print('Model exported to: {}'.format(export_path))
```
